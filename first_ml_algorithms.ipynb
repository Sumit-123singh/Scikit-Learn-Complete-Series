```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features=[[150,0],
         [170,0],
         [130,1],
         [120,1],
         ]
labels=["apple","apple","orange","orange"]
clf=DecisionTreeClassifier()
clf=clf.fit(features,labels)
clf.predict([[150,1],[180,1
                     ]])
```
